## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
regimen_DF = study_data_complete.groupby('Drug Regimen')
# variable to hold the Mean Tumor Volume Data Grouped by Drug Regimen
volume_mean = regimen_DF['Tumor Volume (mm3)'].mean()
# variable to hold median Tumor Volume Data Grouped by Drug Regimen
volume_median = regimen_DF['Tumor Volume (mm3)'].median()
# variable to hold the Tumor Volume Variance Data Grouped by Drug Regimen
volume_var = regimen_DF['Tumor Volume (mm3)'].var()
# variable to hold the Tumor Volume Standard Deviation Data Grouped by Drug Regimen
volume_std = regimen_DF['Tumor Volume (mm3)'].std()
# variable to hold the Tumor Volume SEM Data Grouped by Drug Regimen
volume_sem = regimen_DF['Tumor Volume (mm3)'].sem()
regimen = mice_DF['Drug Regimen'].unique()
# Convert to DataFrame
reg = pd.DataFrame(volume_mean)
reg = reg.rename(columns = {"Tumor Volume (mm3)":'Mean Tumor Volume'})
reg['Median Tumor Volume'] = volume_median
reg['Tumor Volume Variance'] = volume_var
reg['Tumor Volume Std Dev'] = volume_std
reg['Tumor Volume SEM'] = volume_sem
reg
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
#establish x and y values and find St. Pearson Correlation Coefficient for Mouse Weight and Tumor Volume Avg
corrplot = study_data_complete.loc[study_data_complete['Drug Regimen']=="Capomulin"]
corrplot = corrplot.groupby('Mouse ID').mean()
x_value = corrplot['Weight (g)']
y_value = corrplot['Tumor Volume (mm3)']
#print St. Pearson Correlation Coefficient
correlation = st.pearsonr(x_value,y_value)
print(f"The correlation between both factors is {round(correlation[0],2)}")
# establish linear regression values
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_value, y_value)
# linear regression line 
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
regress_values = x_value * slope + intercept
# scatter plot of the data
plt.scatter(x_value,y_value)
plt.plot(x_value,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Tumor Volume Average')
plt.show()
```
